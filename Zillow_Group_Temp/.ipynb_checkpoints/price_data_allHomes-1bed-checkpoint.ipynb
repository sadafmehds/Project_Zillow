```python
csv_path='raw_data/Zip_Zhvi_1bedroom.csv'
df_1b=pd.DataFrame.from_csv(csv_path)

df_1b=df_1b.reset_index()
df_1b=df_1b.set_index('State')
# df_1b

df_1b=df_1b.rename(columns={'RegionName':'Zipcode'})
# df_1b
# len(df_1b)

df_1b=df_1b.loc['CA',:]
df_1b.head()
```

```python
df_1b['Bedrooms']='1'
```

```python
#take needed columns and order
subset=['RegionID','Bedrooms']
mask=df_1b.columns.isin(subset)
df_1b=pd.concat([df_1b.loc[:,mask],df_1b.loc[:,~mask]],axis=1)
```

```python
df_1b=df_1b.loc[:,df_1b.columns!='RegionID']
df_1b
```

```python
df_1b=df_1b.reset_index()
df_1b=df_1b.set_index('CountyName')
df_1b=df_1b.loc['Orange',:]
df_1b
```
