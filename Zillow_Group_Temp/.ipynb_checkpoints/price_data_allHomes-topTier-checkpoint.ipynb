```python
import os
import pandas as pd
import numpy as np
import csv

import seaborn
```

```python
csv_path='raw_data/Zip_Zhvi_TopTier.csv'
df_tt=pd.DataFrame.from_csv(csv_path)

df_tt=df_tt.reset_index()
df_tt=df_tt.set_index('State')
df_tt.head()

df_tt=df_tt.rename(columns={'RegionName':'Zipcode'})
df_tt
len(df_tt)

df_tt=df_tt.loc['CA',:]
df_tt.head()
```

```python
df_tt['Bedrooms']='tt'
```

```python
df_tt
```

```python
#take needed columns and order
subset=['RegionID','Bedrooms']
mask=df_tt.columns.isin(subset)
df_tt=pd.concat([df_tt.loc[:,mask],df_tt.loc[:,~mask]],axis=1)
```

```python
df_tt=df_tt.loc[:,df_tt.columns!='RegionID']
df_tt
```

```python
df_tt=df_tt.reset_index()
df_tt=df_tt.set_index('CountyName')
df_tt=df_tt.loc['Orange',:]
df_tt
```

```python
tt_b_df=pd.DataFrame(df_tt.groupby(['City','Zipcode'])['2009-01','2010-01','2011-01','2012-01','2013-01','2014-01','2015-01','2016-01','2017-01','2018-01'].median())
tt_b_df.head()
```

```python
tt_b_df['Median Price']=tt_b_df.median(axis=1)
tt_b_df.head()


```

```python
tt_b_df['Type']='TopTier'
tt_b_df.head()
```

```python
tt_b_df=tt_b_df.rename(columns={'2009-01':'2009','2010-01':'2010','2011-01':'2011','2012-01':'2012','2013-01':'2013','2014-01':'2014','2015-01':'2015','2016-01':'2016','2017-01':'2017','2018-01':'2018'})
tt_b_df
```
