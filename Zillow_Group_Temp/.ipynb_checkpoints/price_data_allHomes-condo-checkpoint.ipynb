```python
import os
import pandas as pd
import numpy as np
import csv

import seaborn
```

```python
csv_path='raw_data/Zip_Zhvi_Condominum.csv'
df_condo=pd.DataFrame.from_csv(csv_path)

df_condo=df_condo.reset_index()
df_condo=df_condo.set_index('State')
df_condo.head()

df_condo=df_condo.rename(columns={'RegionName':'Zipcode'})
df_condo
len(df_condo)

df_condo=df_condo.loc['CA',:]
df_condo.head()
```

```python
df_condo['Bedrooms']='condo'
```

```python
df_condo
```

```python
#take needed columns and order
subset=['RegionID','Bedrooms']
mask=df_condo.columns.isin(subset)
df_condo=pd.concat([df_condo.loc[:,mask],df_condo.loc[:,~mask]],axis=1)
```

```python
df_condo=df_condo.loc[:,df_condo.columns!='RegionID']
df_condo
```

```python
df_condo=df_condo.reset_index()
df_condo=df_condo.set_index('CountyName')
df_condo=df_condo.loc['Orange',:]
df_condo
```

```python
condo_b_df=pd.DataFrame(df_condo.groupby(['City','Zipcode'])['2009-01','2010-01','2011-01','2012-01','2013-01','2014-01','2015-01','2016-01','2017-01','2018-01'].median())
condo_b_df.head()
```

```python
condo_b_df['Median Price']=condo_b_df.median(axis=1)
condo_b_df.head()


```

```python
condo_b_df['Type']='CONDO'
condo_b_df.head()
```

```python
condo_b_df=condo_b_df.rename(columns={'2009-01':'2009','2010-01':'2010','2011-01':'2011','2012-01':'2012','2013-01':'2013','2014-01':'2014','2015-01':'2015','2016-01':'2016','2017-01':'2017','2018-01':'2018'})
condo_b_df
```
