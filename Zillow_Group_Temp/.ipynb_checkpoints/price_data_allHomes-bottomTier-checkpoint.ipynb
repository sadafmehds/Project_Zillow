```python
import os
import pandas as pd
import numpy as np
import csv

import seaborn
```

```python
csv_path='raw_data/Zip_Zhvi_BottomTier.csv'
df_bt=pd.DataFrame.from_csv(csv_path)

df_bt=df_bt.reset_index()
df_bt=df_bt.set_index('State')
df_bt.head()

df_bt=df_bt.rename(columns={'RegionName':'Zipcode'})
df_bt
len(df_bt)

df_bt=df_bt.loc['CA',:]
df_bt.head()
```

```python
df_bt['Bedrooms']='bottomTier'
```

```python
df_bt
```

```python
#take needed columns and order
subset=['RegionID','Bedrooms']
mask=df_bt.columns.isin(subset)
df_bt=pd.concat([df_bt.loc[:,mask],df_bt.loc[:,~mask]],axis=1)
```

```python
df_bt=df_bt.loc[:,df_bt.columns!='RegionID']
df_bt
```

```python
df_bt=df_bt.reset_index()
df_bt=df_bt.set_index('CountyName')
df_bt=df_bt.loc['Orange',:]
df_bt
```

```python
bt_b_df=pd.DataFrame(df_bt.groupby(['City','Zipcode'])['2009-01','2010-01','2011-01','2012-01','2013-01','2014-01','2015-01','2016-01','2017-01','2018-01'].median())
bt_b_df.head()
```

```python
bt_b_df['Median Price']=bt_b_df.median(axis=1)
bt_b_df.head()


```

```python
bt_b_df['Type']='BottomTier'
bt_b_df.head()
```

```python
bt_b_df=bt_b_df.rename(columns={'2009-01':'2009','2010-01':'2010','2011-01':'2011','2012-01':'2012','2013-01':'2013','2014-01':'2014','2015-01':'2015','2016-01':'2016','2017-01':'2017','2018-01':'2018'})
bt_b_df
```
