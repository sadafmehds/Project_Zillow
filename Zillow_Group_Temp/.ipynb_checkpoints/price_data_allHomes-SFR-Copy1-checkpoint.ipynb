```python
csv_path='raw_data/Zip_Zhvi_SingleFamilyResidence.csv'
df_sfr=pd.DataFrame.from_csv(csv_path)

df_sfr=df_sfr.reset_index()
df_sfr=df_sfr.set_index('State')
df_sfr.head()

df_sfr=df_sfr.rename(columns={'RegionName':'Zipcode'})
df_sfr
len(df_sfr)

df_sfr=df_sfr.loc['CA',:]
df_sfr.head()
```

```python
df_sfr['Bedrooms']='sfr'
```

```python
#take needed columns and order
subset=['RegionID','Bedrooms']
mask=df_sfr.columns.isin(subset)
df_sfr=pd.concat([df_sfr.loc[:,mask],df_sfr.loc[:,~mask]],axis=1)
```

```python
df_sfr=df_sfr.loc[:,df_sfr.columns!='RegionID']
df_sfr
```
