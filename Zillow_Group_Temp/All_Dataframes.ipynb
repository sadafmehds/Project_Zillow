```python
property_crime_07 = crime_07_df[['City','Property crime']]
property_crime_08 = crime_08_df[['City','Property crime']]
property_crime_09 = crime_09_df[['City','Property crime']]
property_crime_10 = crime_10_df[['City','Property crime']]
property_crime_11 = crime_11_df[['City','Property crime']]
property_crime_12 = crime_12_df[['City','Property crime']]
property_crime_13 = crime_13_df[['City','Property crime']]
property_crime_14 = crime_14_df[['City','Property crime']]
property_crime_15 = crime_15_df[['City','Property crime']]
property_crime_16 = crime_16_df[['City','Property crime']]

property_crime_11
```

```python
property_crime_merged = pd.merge(property_crime_07, property_crime_08, on='City', how='inner')
property_crime_merged.rename(columns={'Property crime_x':'2007 Property Crime', 'Property crime_y': '2008 Property Crime'}, inplace=True)

property_crime_merged = pd.merge(property_crime_merged, property_crime_09, on='City', how='inner')
property_crime_merged.rename(columns={'Property crime': '2009 Property crime'}, inplace=True)

property_crime_merged = pd.merge(property_crime_merged, property_crime_10, on='City', how='inner')
property_crime_merged.rename(columns={'Property crime': '2010 Property crime'}, inplace=True)

property_crime_merged = pd.merge(property_crime_merged, property_crime_11, on='City', how='inner')
property_crime_merged.rename(columns={'Property crime': '2011 Property crime'}, inplace=True)

property_crime_merged = pd.merge(property_crime_merged, property_crime_12, on='City', how='inner')
property_crime_merged.rename(columns={'Property crime': '2012 Property crime'}, inplace=True)

property_crime_merged = pd.merge(property_crime_merged, property_crime_13, on='City', how='inner')
property_crime_merged.rename(columns={'Property crime': '2013 Property crime'}, inplace=True)

property_crime_merged = pd.merge(property_crime_merged, property_crime_14, on='City', how='inner')
property_crime_merged.rename(columns={'Property crime': '2014 Property crime'}, inplace=True)

property_crime_merged = pd.merge(property_crime_merged, property_crime_15, on='City', how='inner')
property_crime_merged.rename(columns={'Property crime': '2015 Property crime'}, inplace=True)

property_crime_merged = pd.merge(property_crime_merged, property_crime_16, on='City', how='inner')
property_crime_merged.rename(columns={'Property crime': '2016 Property crime'}, inplace=True)

median_property_crime = property_crime_merged.groupby('City').median()
median = median_property_crime.median(axis=1)

reset_median = median.reset_index()
reset_median.rename(columns={0: 'Median Property Crime'}, inplace=True)
```
